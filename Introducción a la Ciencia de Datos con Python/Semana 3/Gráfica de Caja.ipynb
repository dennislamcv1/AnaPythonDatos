# Gráficas de Caja

<p style='text-align: justify;'>Las <font color = green>__gráficas de caja__</font> son una conveniente representación gráfica de un grupo de datos numéricos utilizando sus cuartiles como referencias. Este tipo de gráficas son utilizadas en la estadística descriptiva y son una excelente forma de analizar datos rápidamente con la ventaja de ocupar un menor espacio, lo cual es útil al comparar distribuciones entre muchos conjuntos de datos.</p>

<p style='text-align: justify;'>Las líneas que se extienden paralelas a las cajas, comúnmente llamadas bigotes, se utilizan para indicar la variabilidad fuera del intervalo formado entre el cuartil inferior y el superior. Los valores atípicos se representan como puntos individuales alineados a los bigotes.</p>

```python
df = pd.read_csv('data/heart/heart.csv')
df.head()
```

```python
import matplotlib.pyplot  as plt
%matplotlib inline
df.boxplot('chol')
plt.title("Boxplot with Chol")
plt.show()
```

```python
import matplotlib.pyplot  as plt
df.boxplot(by = 'sex',column = 'chol')
```

```python
import seaborn as sns
sns.boxplot(x = 'sex', y = 'chol', data = df) 
```

```python
sns.set()
sns.boxplot(x = 'sex', y = 'chol', data = df, width=0.2) 
```

```python
sns.swarmplot(x = 'sex', y = 'chol', data = df, alpha=0.5)
sns.boxplot(x = 'sex', y = 'chol', data = df, width=0.2)
```

```python
sns.boxplot(x = 'sex', y = 'chol', data = df, width=0.2)
sns.swarmplot(x = 'sex', y = 'chol', data = df, alpha=0.5)
plt.title("Colesterol por sexo")
plt.savefig('boxplot.png',dpi=100)
```
